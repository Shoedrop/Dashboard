## Using visualizations to explore patterns in the data

### Host Profile Analysis

```python
sns.pairplot(df[['host_response_rate', 'host_acceptance_rate', 'host_is_superhost', 'review_scores_rating']], hue='host_is_superhost')
plt.show()
```

### Exploration of the Amenities 

```python
# Extract amenities and count their occurrences
amenities = df['amenities'].str.replace('[{}"]', '').str.split(',')
amenities_counts = amenities.explode().value_counts()

# Display the top 10 amenities
amenities_counts.head(10).plot(kind='bar')
plt.title('Top 10 Amenities in Highest-Rated Listings')
plt.ylabel('Count')
plt.show()
```

### Average Price by Neighborhood and Room Type

```python
# Group data by neighborhood and room type and calculate average price
price_optimization = df.groupby(['neighbourhood_cleansed', 'room_type'])['price'].mean().reset_index()

# Visualizing the average price for room types across neighborhoods
sns.barplot(x='neighbourhood_cleansed', y='price', hue='room_type', data=price_optimization)
plt.xticks(rotation=90)
plt.title('Average Price by Neighborhood and Room Type')
plt.show()
```

### Distributions of Numerical Variables

```python
# Selecting a few numerical columns for distribution plots
numerical_columns = ['host_listings_count', 'accommodates', 'bedrooms', 'price', 'number_of_reviews', 'reviews_per_month']

# Plotting distributions of numerical columns
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
fig.suptitle('Distributions of Numerical Variables', fontsize=16)

for col, ax in zip(numerical_columns, axes.flatten()):
    sns.histplot(df[col], bins=30, ax=ax, kde=True)
    ax.set_title(col.replace('_', ' ').title())
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

### Counts of Categorical Variables

```python
# Showing counts of a few categorical variables
categorical_columns = ['room_type', 'instant_bookable']

# Plotting counts of categorical columns
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
fig.suptitle('Counts of Categorical Variables', fontsize=16)

for col, ax in zip(categorical_columns, axes.flatten()):
    sns.countplot(x=df[col], ax=ax)
    ax.set_title(col.replace('_', ' ').title())
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.grid(True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

### Correlation Heatmap of Numerical Variables

```python
# Correlation heatmap between the different variables 
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
